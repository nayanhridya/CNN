```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image
image = cv2.imread('/content/flower.jpeg')

# Convert the image to RGB (from BGR, which OpenCV uses)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert to grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Step 3: Define the convolutional kernels

# Original kernel (vertical edge detection)
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

# Sobel Kernel (Horizontal Edge Detection)
sobel_horizontal = np.array([[-1, -2, -1],
                             [ 0,  0,  0],
                             [ 1,  2,  1]])

# Gaussian Blur Kernel
gaussian_blur = np.array([[1,  4,  6,  4, 1],
                          [4, 16, 24, 16, 4],
                          [6, 24, 36, 24, 6],
                          [4, 16, 24, 16, 4],
                          [1,  4,  6,  4, 1]]) / 256

# Sharpen Kernel
sharpen = np.array([[ 0, -1,  0],
                    [-1,  5, -1],
                    [ 0, -1,  0]])

# Step 4: Apply the convolution operation using each kernel
convolved_image_1 = cv2.filter2D(image_gray, -1, kernel)
convolved_image_2 = cv2.filter2D(image_gray, -1, sobel_horizontal)
convolved_image_3 = cv2.filter2D(image_gray, -1, gaussian_blur)
convolved_image_4 = cv2.filter2D(image_gray, -1, sharpen)

# Step 5: Visualize the original image, kernels, and convolved images
plt.figure(figsize=(20, 10))

plt.subplot(2, 4, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(2, 4, 2)
plt.title("Vertical Edge Kernel")
plt.imshow(kernel, cmap='gray', interpolation='none')
plt.axis('off')

plt.subplot(2, 4, 3)
plt.title("Sobel Kernel")
plt.imshow(sobel_horizontal, cmap='gray', interpolation='none')
plt.axis('off')

plt.subplot(2, 4, 4)
plt.title("Gaussian Blur Kernel")
plt.imshow(gaussian_blur, cmap='gray', interpolation='none')
plt.axis('off')

plt.subplot(2, 4, 5)
plt.title("Vertical Edge Detection")
plt.imshow(convolved_image_1, cmap='gray')
plt.axis('off')

plt.subplot(2, 4, 6)
plt.title("Sobel Horizontal Edge")
plt.imshow(convolved_image_2, cmap='gray')
plt.axis('off')

plt.subplot(2, 4, 7)
plt.title("Gaussian Blur Effect")
plt.imshow(convolved_image_3, cmap='gray')
plt.axis('off')

plt.subplot(2, 4, 8)
plt.title("Sharpen Effect")
plt.imshow(convolved_image_4, cmap='gray')
plt.axis('off')

plt.show()

```
